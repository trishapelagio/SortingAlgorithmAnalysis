# 1. Pi Approximation

## Variant 1

```python
def pi_var1(r):
    inCount = 0
    rsquare = r*r
    ctr = 0
    for x in range(1, r+1):
        xsquare = x*x
        for y in range(1, r+1):
            if xsquare + y*y <= rsquare:
                inCount += 1
    return 4*inCount/rsquare
```

```python
N = [100,400,900,1600,2500,3600,4900,6400,8100,10000]
piv1_count = []
for i in N:
    piv1_count.append([i, pi_var1(int(math.sqrt(i)))])
piv1_count = pd.DataFrame(piv1_count, columns=['N', 'Pi'])
piv1_count.plot(x='N', y='Pi', label='Variation 1')
plt.xlabel("N")
plt.ylabel("Pi")
```

## Variant 2

```python
def pi_var2(N):
    inCount = 0
    for i in range(1, N+1):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        if x*x + y*y <= 1:
            inCount += 1
    return 4 * inCount / N
```

```python
N = [100,400,900,1600,2500,3600,4900,6400,8100,10000]
piv2_count = []
for i in N:
    piv2_count.append([i, pi_var2(i)])
piv2_count = pd.DataFrame(piv2_count, columns=['N', 'Pi'])
piv2_count.plot(x='N', y='Pi', label='Variation 2')
plt.xlabel("N")
plt.ylabel("Pi")
```

## Variant 3

```python
def pi_var3(r):
    inCount = 0
    ctr = 0
    for xInt in range(1, r+1):
        x = xInt/r
        xsquare=x*x
        for yInt in range(1, r+1):
            y = yInt/r
            if xsquare+y*y <= 1:
                inCount += 1
    return 4 * inCount / (r*r)
```

```python
N = [100,400,900,1600,2500,3600,4900,6400,8100,10000]
piv3_count = []
for i in N:
    piv3_count.append([i, pi_var3(int(math.sqrt(i)))])
piv3_count = pd.DataFrame(piv3_count, columns=['N', 'Pi'])
piv3_count.plot(x='N', y='Pi', label='Variation 3')
plt.xlabel("N")
plt.ylabel("Pi")
```

```python
fibo = 0

def fibonacci(n):
    global fibo
    fibo += 1

    if n == 1 or n == 2:
        return 1

    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
N = [100,200,300,400,500,600,700,800,900,1000]
arrays = []
for i in N:
    array = []
    for j in range(i):
        array.append(j+1)
    arrays.append(array)
```

```python
def linearSearch(arr, n, x):
    global linearSearch_ctr
    for i in range(0, n):
        linearSearch_ctr += 1;
        if (arr[i] == x):
            return i
    return -1
```

```python
lin_count = []
for i in arrays:
    ave = 0
    for j in range(5):
        linearSearch_ctr = 0
        x = random.randint(1,len(i))
        linearSearch(i, len(i), x)
        ave += linearSearch_ctr
    lin_count.append([len(i), ave/5])
lin_count = pd.DataFrame(lin_count, columns=['N', 'Frequency'])
lin_count.plot(x='N', y='Frequency')
```

## Binary Search Function

```python
def binarySearch(arr, l, r, x):
    global binarySearch_ctr
    binarySearch_ctr += 1
    if r >= l:
        mid = l + (r - l) // 2
        if arr[mid] == x:
            return mid

        elif arr[mid] > x:
            return binarySearch(arr, l, mid - 1, x)

        else:
            return binarySearch(arr, mid + 1, r, x)

    else:
        return -1
```

```python
bin_count = []
for i in arrays:
    ave = 0
    for j in range(5):
        binarySearch_ctr = 0
        x = random.randint(1,len(i))
        binarySearch(i, 0, len(i), x)
        ave += binarySearch_ctr
    bin_count.append([len(i), ave/5])
bin_count = pd.DataFrame(bin_count, columns=['N', 'Frequency'])
bin_count.plot(x='N', y='Frequency', label="Binary Count")
plt.xlabel("N")
plt.ylabel("Frequency")
```

## Insertion Sort

```python
def insertionSort(numbers, N):
    freq = 0
    for j in range(1, N):
        key = numbers[j]
        i = j-1
        while i >= 0 and numbers[i] > key:
            freq += 1
            numbers[i+1] = numbers[i]
            i = i-1
        freq += 1
        numbers[i+1] = key
    return numbers, freq
```

## Bubble Sort

```python
def bubbleSort(numbers, N):
    freq = 0
    i = N
    while i >= 0:
        for j in range(1, i):
            freq += 1
            if numbers[j-1] > numbers[j]:
                numbers[j-1], numbers[j] = numbers[j], numbers[j-1]
        i -= 1
    return numbers, freq
```

## Selection Sort

```python
def selectionSort(numbers, N):
    freq = 0
    for i in range(N):
        min = i
        for j in range(i+1, N):
            freq += 1
            if numbers[j] < numbers[min]:
                min = j

        numbers[min], numbers[i] = numbers[i], numbers[min]

    return numbers, freq
```

```python
# def partition(arr, low, high):
#     i = (low - 1)  # index of smaller element
#     pivot = arr[high]  # pivot
#     global quickSort_ctr

#     for j in range(low, high):

#         # If current element is smaller than or
#         # equal to pivot
#         quickSort_ctr += 1
#         if arr[j] <= pivot:
#             # increment index of smaller element
#             i = i + 1
#             arr[i], arr[j] = arr[j], arr[i]

#     arr[i + 1], arr[high] = arr[high], arr[i + 1]
#     return (i + 1)


# def quickSort(arr, low, high):
#     if len(arr) == 1:
#         return arr
#     if low < high:
#         # pi is partitioning index, arr[p] is now
#         # at right place
#         pi = partition(arr, low, high)

#         # Separately sort elements before
#         # partition and after partition
#         quickSort(arr, low, pi - 1)
#         quickSort(arr, pi + 1, high)
#     return arr

## Pivot at start
def partition(array, start, end):
    global quickSort_ctr
    pivot = array[start]
    low = start + 1
    high = end

    while True:
        # If the current value we're looking at is larger than the pivot
        # it's in the right place (right side of pivot) and we can move left,
        # to the next element.
        # We also need to make sure we haven't surpassed the low pointer, since that
        # indicates we have already moved all the elements to their correct side of the pivot
        while low <= high and array[high] >= pivot:
            quickSort_ctr += 1
            high = high - 1

        # Opposite process of the one above
        while low <= high and array[low] <= pivot:
            quickSort_ctr += 1
            low = low + 1

        # We either found a value for both high and low that is out of order
        # or low is higher than high, in which case we exit the loop
        if low <= high:
            array[low], array[high] = array[high], array[low]
            # The loop continues
        else:
            # We exit out of the loop
            break

    array[start], array[high] = array[high], array[start]

    return high

def quickSort(array, start, end):
    if start >= end:
        return

    p = partition(array, start, end)
    quickSort(array, start, p-1)
    quickSort(array, p+1, end)
```

## Merge Sort

```python
def merge(arr, l, m, r):
    global mergeSort_ctr
    n1 = m - l + 1
    n2 = r - m

    # create temp arrays
    L = [0] * (n1)
    R = [0] * (n2)

    # Copy data to temp arrays L[] and R[]
    for i in range(0, n1):
        L[i] = arr[l + i]

    for j in range(0, n2):
        R[j] = arr[m + 1 + j]

    # Merge the temp arrays back into arr[l..r]
    i = 0  # Initial index of first subarray
    j = 0  # Initial index of second subarray
    k = l  # Initial index of merged subarray

    while i < n1 and j < n2:
        mergeSort_ctr += 1
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1

    # Copy the remaining elements of L[], if there
    # are any
    while i < n1:
        mergeSort_ctr += 1
        arr[k] = L[i]
        i += 1
        k += 1

    # Copy the remaining elements of R[], if there
    # are any
    while j < n2:
        mergeSort_ctr += 1
        arr[k] = R[j]
        j += 1
        k += 1


def mergeSort(arr, l, r):
    if l < r:
        # Same as (l+r)//2, but avoids overflow for
        # large l and h
        m = (l + (r - 1)) // 2

        # Sort first and second halves
        mergeSort(arr, l, m)
        mergeSort(arr, m + 1, r)
        merge(arr, l, m, r)
    return arr
```

## Shell Sort

```python
def shellSort(alist):
    sublistcount = len(alist)//2
    while sublistcount > 0:
      for start_position in range(sublistcount):
        gap_InsertionSort(alist, start_position, sublistcount)

      sublistcount = sublistcount // 2
    return alist


def gap_InsertionSort(nlist,start,gap):
    global shellSort_ctr
    for i in range(start+gap,len(nlist),gap):

        current_value = nlist[i]
        position = i

        while position>=gap and nlist[position-gap]>current_value:
            shellSort_ctr += 1
            nlist[position]=nlist[position-gap]
            position = position-gap
        shellSort_ctr += 1
        nlist[position]=current_value
```
